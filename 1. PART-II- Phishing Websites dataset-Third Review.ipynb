You need to know about the distribution of the websites so you understand what percentage of websites are phishing in 
the dataset. Use the Result column of the dataset and the CounterPython module to know the class distribution of the 
websites. After getting the numbers, arrange and convert them to a pandas DataFrame, which should resemble the following.

```python
# import Counter from collections 
from collections import Counter 
  
# creating a raw data-set using keyword arguments 
x = Counter(data['Result'])

#Indicate the elements
e = x.elements
  
# printing out the elements 
print(e)
```

```python
for i in x: 
    print(e)
```

```python
classes = Counter(data["Result"].values)
classes.most_common()
```

```python
class_dist = pd.DataFrame(classes.most_common(), columns=['Class', 'Num_Observations'])
class_dist
```

```python
import pandas as pd

url="./Phishing.csv"
df=pd.read_csv(url)

value_counts=df.Result.value_counts()
df2 = value_counts.rename_axis('Class').reset_index(name='Num_Observations')
```

```python
import matplotlib.pyplot as plt

# Stylize the plots
plt.style.use("ggplot")

colors = ["r", "b"]
ax = df2.groupby("Class")["Num_Observations"].sum().plot(kind="barh", width=0.2, figsize=(10,8), color=colors)

ax.set_title("Class distribution of the websites", fontsize = 15)
ax.set_xlabel("Class", fontsize=12)
ax.set_ylabel("Nuber of Observations", fontsize=12)

for i in ax.patches:
    ax.text(i.get_width()+0.1, i.get_y()+0.1, \
            str(i.get_width()), fontsize=11)

plt.show()

df.describe().T
```
